```python
import numpy as np
import pandas as pd

import seaborn as sns

%matplotlib inline

import matplotlib as mpl 
import matplotlib.pyplot as plt

import plotly
import plotly.graph_objects as go 
```

```python
vacs_world = pd.read_csv(r'C:\Users\patka\enterprise_data_science-covid-19\data\COVID-19-master\data_tables\vaccine_data\global_data\time_series_covid19_vaccine_global.csv')

vacs_world.head()
```

We can eliminate Province_State and Date coloumn

Considering Report_Date_String coloumn only and renaming the remaining coloumns for clarity

```python
vacs_raw = vacs_world.drop(['Date','Province_State'], axis=1)

vacs_raw.head()
```

```python
vacs_raw = vacs_raw.rename(columns = {"Country_Region":"Country","People_partially_vaccinated":"Partial_Vax", "People_fully_vaccinated":"Full_Vax", "Report_Date_String":"Date" })

vacs_raw.head()
```

```python
#Getting the date in string format for ease in representation and plotting

vacs_raw['Date'] = pd.to_datetime(vacs_raw['Date'])
```

```python
#Implementing Pivot-Table
pivot = pd.pivot_table(data = vacs_raw,
                       index = 'Date',
                       columns = 'Country',
                       values = 'Full_Vax',
                      )

pivot = pivot.fillna(0)
```

```python
pivot = pivot.replace(np.nan,0)

pivot.reset_index(inplace = True)

pivot.head()
```

```python
time_idx= pivot["Date"]

df_plot = pd.DataFrame({'Date': time_idx})

df_plot.head()
```

```python
country_list1=['Italy', 'US', 'Spain','India','Germany']
```

```python
wp_country = pd.read_csv(r'C:\Users\patka\enterprise_data_science-covid-19\data\COVID-19-master\data_tables\world_pop_by_country.csv')

wp_country.head()
```

```python
for each in country_list1:
    df_plot[each] = pivot[each]
    var = np.array(wp_country.loc[wp_country["Country Name"]==each, "2018"])
    df_plot[each] = df_plot[each]/var[0] 
    
df_plot.head()
```
